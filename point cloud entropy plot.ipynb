```python
import os
import pandas as pd
import plotly.graph_objects as go

# Parent directory containing folders with Excel files for each image
# entropy_parent_folder = "C:\\Users\\Aneesh PB\\Downloads\\Topology Project\\Full segment\\Entropy\\Normal cases\\"

# entropy_parent_folder = "C:\\Users\\Aneesh PB\\Downloads\\Topology Project\\VS code\\1000points\\Entropy\\Normal cases\\"

# "C:\Users\Aneesh PB\Downloads\Topology Project\VS code\1000points\Entropy\Malignant cases\Malignant case (99)_Entropy_values.xlsx"
entropy_parent_folder = "C:\\Users\\Aneesh PB\\Downloads\\Topology Project\\Full segment\\Entropy code\\500\\Entropy\\Normal cases\\"
# Initialize dictionaries to store entropy values for each x value
entropy_0_dict = {}
entropy_1_dict = {}

# Iterate over all the entropy files
for entropy_file in os.listdir(entropy_parent_folder):
    if entropy_file.endswith('_Entropy_values.xlsx'):
        entropy_file_path = os.path.join(entropy_parent_folder, entropy_file)

        # Read the entropy values DataFrame
        entropy_df = pd.read_excel(entropy_file_path)

        # Extract x values from the first column
        x_values = entropy_df.iloc[:, 0]#.apply(lambda x: int(x.split('.')[0]))

        # Extract numerical part from the values in arrays
        # entropy_df['Entropy 0'] = entropy_df['Entropy 0'].apply(lambda x: float(x[1:-1]))
        # entropy_df['Entropy 1'] = entropy_df['Entropy 1'].apply(lambda x: float(x[1:-1]))

        # Update entropy dictionaries with mean values for each x value
        for x, entropy_0, entropy_1 in zip(x_values, entropy_df['Entropy 0'], entropy_df['Entropy 1']):
            if x not in entropy_0_dict:
                entropy_0_dict[x] = [entropy_0]
                entropy_1_dict[x] = [entropy_1]
            else:
                entropy_0_dict[x].append(entropy_0)
                entropy_1_dict[x].append(entropy_1)

# Calculate mean entropy values for each x value
mean_entropy_0 = {x: sum(values) / len(values) for x, values in entropy_0_dict.items()}
mean_entropy_1 = {x: sum(values) / len(values) for x, values in entropy_1_dict.items()}
mean_entropy_0 = dict(sorted(mean_entropy_0.items()))
mean_entropy_1 = dict(sorted(mean_entropy_1.items()))

# Create plots for mean entropy 0 and mean entropy 1
fig0 = go.Figure()
fig0.add_trace(go.Scatter(x=list(mean_entropy_0.keys()), y=list(mean_entropy_0.values()),
                           mode='lines', name='Mean Entropy 0', line=dict(color='blue')))
fig0.update_layout(title='Mean Entropy 0 for All Images',
                   xaxis_title='X Values',
                   yaxis_title='Mean Entropy Value')

fig1 = go.Figure()
fig1.add_trace(go.Scatter(x=list(mean_entropy_1.keys()), y=list(mean_entropy_1.values()),
                           mode='lines', name='Mean Entropy 1', line=dict(color='red')))
fig1.update_layout(title='Mean Entropy 1 for All Images',
                   xaxis_title='X Values',
                   yaxis_title='Mean Entropy Value')

fig0.show()
fig1.show()
```
