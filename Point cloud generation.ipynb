```python
#final

import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def process_image(filename_image, output_parent_folder):
    try:
        # Load the grayscale image
        inputImage = plt.imread(filename_image)

        # Check if the image is in color (3 channels)
        if inputImage.ndim == 3:
            # Convert to grayscale
            inputImage = np.mean(inputImage, axis=-1)

        # Get the dimensions of the image
        rows, cols = inputImage.shape

        # Create x and y coordinate grids
        x, y = np.meshgrid(np.arange(cols), np.arange(rows))

        # Reshape the x, y, and intensity arrays
        x = x.flatten()
        y = y.flatten()
        intensity = inputImage.flatten()

        # Extract image name without extension
        image_name = os.path.splitext(os.path.basename(filename_image))[0]

        # Create output folder for the image if it doesn't exist
        output_folder = os.path.join(output_parent_folder, image_name)
        os.makedirs(output_folder, exist_ok=True)

        # Write coordinates and intensity values to Excel files for each intensity
        unique_intensities = np.unique(intensity)
        for intensity_value in unique_intensities:
            # Filter points corresponding to the current intensity
            points_with_intensity = [(x[i], y[i]) for i, inten in enumerate(intensity) if inten == intensity_value]

            # Create DataFrame from the filtered points
            df = pd.DataFrame(points_with_intensity, columns=['X', 'Y'])

            # Save DataFrame to Excel file
            excel_filename = os.path.join(output_folder, f'{intensity_value:.0f}.xlsx')
            df.to_excel(excel_filename, index=False)
            # print(f"Excel file saved for image {image_name} and intensity {intensity_value:.0f} as {excel_filename}")

    except Exception as e:
        print(f"Error processing {filename_image}: {e}")


# Folder containing images
input_folder = "C:\\Users\\Aneesh PB\\Downloads\\Topology Project\\Full segment\\Images\\Normal cases"
# Output folder for each image's Excel files
output_parent_folder = "C:\\Users\\Aneesh PB\\Downloads\\Topology Project\\Full segment\\Point Cloud\\Normal cases\\"

i=0
# Process each image in the input folder
for filename in os.listdir(input_folder):
    # if i==20:
    #     break
    # i+=1
    if filename.lower().endswith('.png') or filename.lower().endswith('.jpg'):  # Adjust file extensions as needed
        filename_image = os.path.join(input_folder, filename)
        process_image(filename_image, output_parent_folder)
    print("Finished",filename)
```
