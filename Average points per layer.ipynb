```python
import os
import numpy as np
import matplotlib.pyplot as plt

def process_image_and_plot_intensity_distribution(filename_image):
    try:
        # Load the grayscale image
        inputImage = plt.imread(filename_image)

        # Check if the image is in color (3 channels)
        if inputImage.ndim == 3:
            # Convert to grayscale
            inputImage = np.mean(inputImage, axis=-1)

        # Flatten the intensity array
        intensity = inputImage.flatten()

        # Extract image name without extension
        image_name = os.path.splitext(os.path.basename(filename_image))[0]

        # Count the number of points for each intensity value
        unique_intensities, counts = np.unique(intensity, return_counts=True)

        # Plot the number of points for each intensity value
        plt.figure(figsize=(10, 6))
        plt.bar(unique_intensities, counts, color='blue', alpha=0.7)
        plt.xlabel('Intensity Value')
        plt.ylabel('Number of Points')
        plt.title(f'Number of Points for Each Intensity Value in {image_name}')
        plt.grid(True)
        plt.tight_layout()
        plt.show()

    except Exception as e:
        print(f"Error processing {filename_image}: {e}")

# Folder containing images
input_folder = "C:\\Users\\Aneesh PB\\Downloads\\Topology Project\\New Data\\Images\\Normal cases\\"

# Process each image in the input folder
for filename in os.listdir(input_folder):
    if filename.lower().endswith('.png') or filename.lower().endswith('.jpg'):  # Adjust file extensions as needed
        filename_image = os.path.join(input_folder, filename)
        process_image_and_plot_intensity_distribution(filename_image)
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import pandas as pd

def process_image(filename_image):
    try:
        # Load the image
        inputImage = plt.imread(filename_image)

        # Check if the image is in color (3 channels)
        if inputImage.ndim == 3:
            # Convert to grayscale
            inputImage = np.mean(inputImage, axis=-1)

        # Flatten the intensity array
        intensity = inputImage.flatten()

        return intensity

    except Exception as e:
        print(f"Error processing {filename_image}: {e}")
        return None

# Folder containing images
input_folder = "C:\\Users\\Aneesh PB\\Downloads\\Topology Project\\New Data\\Images\\Normal cases\\"

# Dictionary to store the sum of counts for each intensity value across all images
intensity_count_sum = {}
# Total number of images processed
num_images_processed = 0

# Process each image in the input folder
for filename in os.listdir(input_folder):
    if filename.lower().endswith('.png') or filename.lower().endswith('.jpg'):  # Adjust file extensions as needed
        filename_image = os.path.join(input_folder, filename)
        intensity = process_image(filename_image)
        if intensity is not None:
            unique_intensities, counts = np.unique(intensity, return_counts=True)
            for intensity_value, count in zip(unique_intensities, counts):
                if intensity_value in intensity_count_sum:
                    intensity_count_sum[intensity_value] += count
                else:
                    intensity_count_sum[intensity_value] = count
            num_images_processed += 1

# Calculate the average number of points for each intensity value
average_intensity_counts = {intensity: count / num_images_processed for intensity, count in intensity_count_sum.items()}

# Create a DataFrame for plotting
data = {'Intensity': list(average_intensity_counts.keys()), 'Average Count': list(average_intensity_counts.values())}
df = pd.DataFrame(data)

# Create an interactive plot
fig = px.bar(df, x='Intensity', y='Average Count',
             title='Average Number of Points for Each Intensity Value Across All Images',
             labels={'Intensity': 'Intensity Value', 'Average Count': 'Average Number of Points'})

# Show the plot
fig.show()
```
